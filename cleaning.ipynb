```python
# Dependencies
import pandas as pd
import matplotlib as plt
```

```python
# Import CSVs
base_file = "-citibike-tripdata.csv"

jan_data = pd.read_csv(f"Resources/202101{base_file}")
feb_data = pd.read_csv(f"Resources/202102{base_file}")
mar_data = pd.read_csv(f"Resources/202103{base_file}")
```

```python
# Create First Quarter df
q1_data = pd.concat([jan_data, feb_data, mar_data]).reset_index(drop=True)
q1_df = pd.DataFrame(q1_data)
q1_df.head()
```

```python
# Split the start and end time columns and rename the originals
q1_df[["start date","start time"]] = q1_df["starttime"].str.split(expand=True)
q1_df[["stop date","stop time"]] = q1_df["stoptime"].str.split(expand=True)
q1_df.rename(columns={"starttime": "start datetime", "stoptime": "stop datetime"})
```

```python
# Replace gender numbers
q1_df.replace({"gender":{0:"-",1:"M",2:"F"}})
```

```python
q1_df.to_csv("Resources/citibike_2021_Q1.csv",index=False)
```
